##MODELS INTEGRATION

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
model
```

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response = model.invoke("How hard is German language?")
response.content
```

##STREAMING

```python
model.invoke("Write a 200 words paragraph on Jaipur")
```

```python
for chunk in model.stream("Write a 200 words paragraph on Jaipur"):
    print(chunk.text)
```

```python
for chunk in model.stream("Why does a person need protein in their diet?"):
    print(chunk.text, end="|", flush=True)
```

##BATCH

```python
responses = model.batch([
    "What does palantir do?",
    "What is the jurassic period",
    "How many triangles in a square"
])

for response in responses:
    print(response)
```
